```python
import pandas as pd
import numpy as np
import joblib

print("Libraries imported successfully ")
```

```python
df = pd.read_csv("../dataset/customer_support_data.csv")

print(df.head())
print(df.columns)
print(df.shape)
```

```python
# Convert query column to lowercase for matching
df["Query"] = df["Query"].str.lower()

print(df)
```

```python
def chatbot_response(user_query):
    user_query = user_query.lower()

    for index, row in df.iterrows():
        if user_query == row["Query"]:
            return row["Response"]

    return "Sorry, I could not understand your query. Please contact support."


# Example Test
print(chatbot_response("Where is my order?"))
print(chatbot_response("Refund status"))
```

```python
joblib.dump(df, "../models/chatbot_data.pkl")

print("Chatbot data saved successfully ")
```
